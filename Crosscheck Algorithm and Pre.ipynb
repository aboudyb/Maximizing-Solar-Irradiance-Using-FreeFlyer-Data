```python
#Hour Crosschecker and Preproccesing 
import pandas as pd
import numpy as np

dF=pd.read_csv('C:/Users/aboud/Dissertation/Pre Training Data/PreLongitude_55_1.csv')
xComponents=dF['SunXComponents[index]']
yComponents=dF['SunYComponents[index]']
zComponents=dF['SunZComponents[index]']
TS=dF['Azimuth[index]']
RS=dF['Elevation[index]']
#Spacecraft's body axes vectors
spacecraftYAxis=np.array([0,1,0])
spacecraftXAxis=np.array([1,0,0])

def adjustForAlphaRS(hour, DegRS, hourDegRS):
    #Detect if angle has wrapped around from 180 back to 0, if so map back to 180 - 360
    if DegRS<hourDegRS-180:
        return 360-DegRS
    return DegRS

averagedAzimuths,averagedElevations,squaredErrorsTS,squaredErrorsRS=[],[],[],[]
#Threshold for angle comparison
threshold=np.radians(0.1)

for hour in range(1, len(xComponents)+1):
    #Calculate angles as before
    sunSpacecraftVector=np.array([xComponents[hour-1],yComponents[hour-1],zComponents[hour-1]])
    RadTS=np.arccos(np.clip(np.dot(sunSpacecraftVector,spacecraftXAxis)/np.linalg.norm(sunSpacecraftVector),-1,1))
    RadRS=np.arccos(np.clip(np.dot(np.array([0,yComponents[hour-1],zComponents[hour-1]]),spacecraftYAxis)/np.linalg.norm([0,yComponents[hour-1],zComponents[hour-1]]),-1,1))
    #Convert to degrees for comparison and potential adjustment
    DegTS=np.degrees(RadTS)
    DegRS=np.degrees(RadRS)
    #Adjust the alphaRS angle if necessary
    adjustedDegRS=adjustForAlphaRS(hour, DegRS, RS[hour-1])
    #Compare
    matchTS=abs(DegTS-TS[hour-1])<=np.degrees(threshold)
    matchRS=abs(adjustedDegRS-RS[hour-1])<=np.degrees(threshold)
    #Calculate averages
    averageTSDeg=(DegTS+TS[hour-1])/2
    averageRSDeg=(adjustedDegRS+RS[hour-1])/2
    averagedAzimuths.append(averageTSDeg)
    averagedElevations.append(averageRSDeg)
    squaredErrorsTS.append((DegTS-TS[hour-1])**2)
    squaredErrorsRS.append((adjustedDegRS-RS[hour-1])**2)
    print(f" AlphaTS match: {matchTS}, Calculated: {DegTS}, Simulated: {TS[hour-1]}, Azimuth: {averageTSDeg} degrees")
    print(f" AlphaRS match: {matchRS}, Calculated: {adjustedDegRS}, Simulated: {RS[hour-1]}, Elevation: {averageRSDeg} degrees")
#Calculate MSE and save averages
mseAzimuth=np.mean(squaredErrorsTS)
mseElevation=np.mean(squaredErrorsRS)
averagedAngles=pd.DataFrame({'Azimuth':averagedAzimuths,'Elevation':averagedElevations})
averagedAngles.to_csv('C:/Users/aboud/Dissertation/Training Data/Longitude_55_1.csv', index=False)
mse=pd.DataFrame({'MSE_Azimuth':[mseAzimuth],'MSE_Elevation':[mseElevation]})
mse.to_csv('C:/Users/aboud/Dissertation/MSE of Azimuth and Elevation/MSELongitude_55_1.csv', index=False)

print(f"Mean Squared Error for Azimuth: {mseAzimuth}")
print(f"Mean Squared Error for Elevation: {mseElevation}")
```
